Markdown Cheat Sheet for Jupyter Notebook / VS Code
Markdown is a lightweight markup language that allows you to format text using standard keyboard characters. Your brain processes it linearly (you write code), and the editor compiles it into a clean, readable view.
1. Headers (Note Structuring)
Use the # (hash) symbol. The number of hashes determines the header level:
Level 1 Header (Main Title)
Level 2 Header (Section / Chapter)
Level 3 Header (Subsection / Task)
Level 4 Header (Details)
2. Text Formatting
Bold: Use double asterisks: important text -> important text
Italic: Use single asterisks: *definition* -> definition
Inline code: Use backticks (the key below the tilde ~): `variable_x` -> variable_x
Strikethrough: Use double tildes: ~~wrong assumption~~ -> ~~wrong assumption~~
3. Lists (Bullet points)
Unordered list (bullets):
Use a dash - or an asterisk *:
First step of the algorithm
And the second step!
Subpoint (just press Tab)
Subpoint 2
Ordered list (numbers):
Simply type a number followed by a period:
Variable initialization
Main loop
Rendering the plot
4. Code Blocks
If you want to paste a snippet of Python code (e.g., for reading, without executing it), use triple backticks and specify the language name:
def calculate_velocity(s, t):
    return s / t


5. Mathematics (LaTeX / MathJax)
This is the most important feature for you.
Inline (within a sentence): Use single dollars: The result is $x^2 + y^2 = r^2$ -> The result is 
Display (in a separate block): Use double dollars:
$$\int_{0}^{\infty} e^{-x^2} dx$$
Resizing mathematical formulas (Sizing)
If the default math font size is too small compared to your code, you can force a larger size by adding a specific command right after opening the math block.
Available sizes (from smallest to largest):
\large
\Large
\LARGE
\huge
\Huge

6. Blockquotes
Used to highlight definitions from a textbook or important laws of physics. Use the > sign:
"Mathematics is the alphabet with which God has written the universe." - Galileo Galilei
7. Links and Images (e.g., tablet scans)
Link: [visible text](url_address)
Example: [My repository](https://github.com)
Image: Add an exclamation mark at the beginning: ![alternative description](path/to/file.png)
Example: ![My calculations from Boas](../assets/task14.png)

#### 2.6 
$$\frac{n^n}{n!}$$

```python
import matplotlib.pyplot as plt
import math

# 1. Simulation setup
# We keep the range small (1 to 20) because n^n grows insanely fast.
max_n = 20
n_values = list(range(1, max_n + 1))
results = []

# 2. Main calculation loop
for n in n_values:
    # Calculate the numerator (n to the power of n) and denominator (n factorial)
    numerator = n ** n
    denominator = math.factorial(n)
    
    # Python handles massive integers natively, so we can just divide
    value = numerator / denominator
    results.append(value)

# Print a few values to the console for a quick check
print("Sequence values for the first few steps:")
for i in [1, 2, 5, 10, 19]:
    print(f"For n={i}: the value is approximately {results[i-1]:.2f}")

# 3. Plotting the chart
plt.style.use('dark_background') # High-contrast theme
plt.figure(figsize=(10, 6))

# Draw the main line
plt.plot(n_values, results, marker='o', markersize=6, color='#FF5733', linewidth=2, label='a_n = n^n / n!')

# Aesthetics and labels
plt.title('The graph of: n^n vs n!', fontsize=14, pad=15)
plt.xlabel('Step (n)', fontsize=12)
plt.ylabel('Sequence Value', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.3)
plt.legend(fontsize=12)

# Render the plot
plt.show()
```

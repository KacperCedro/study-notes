```python
import matplotlib.pyplot as plt
import numpy as np # Dodajemy Numpy!

# 1. Przygotowanie danych (np. badamy ułamek n^2 / 2^n)
# Zamiast range(), używamy np.arange(start, stop, krok)
n_values = np.arange(1, 10.1, 0.1) 

# Magia Numpy (Wektoryzacja) - koniec z pętlą for!
liczniki = n_values**2
mianowniki = 2**n_values
wartosci_calkowite = liczniki / mianowniki

# 2. Tworzenie płótna
plt.figure(figsize=(10, 6))

# 3. RYSOWANIE WIELU LINII (Wersja dostępna i do druku B&W)
# markevery=10 -> marker pojawia się co 10 punktów (brak zlewania się)
# linestyle -> unikalny styl linii (ciągła '-', przerywana '--', kropkowana ':')
plt.plot(n_values, liczniki, color='blue', linestyle='-', marker='o', markevery=10, linewidth=2, label='Licznik ($n^2$)')
plt.plot(n_values, mianowniki, color='red', linestyle='--', marker='s', markevery=10, linewidth=2, label='Mianownik ($2^n$)')
plt.plot(n_values, wartosci_calkowite, color='green', linestyle=':', marker='^', markevery=10, linewidth=3, label='Wyrażenie (L/M)')

# 4. Estetyka i opisy
plt.title('Porównanie licznika, mianownika i całego ułamka', fontsize=14, pad=15)
plt.xlabel('Kroki (n)', fontsize=12)
plt.ylabel('Wartość', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)

# KLUCZOWE: Włączenie legendy
plt.legend(fontsize=12, loc='upper left')

# 5. Wyświetlenie całości
plt.show()
```

```python
import numpy as np
import pandas as pd
from IPython.display import display
from IPython.display import Markdown

# 1. Generate numerical data
n_values = np.arange(1, 11)
numerators = n_values ** 2
denominators = 2 ** n_values
results = numerators / denominators

# 2. Create Dictionary with LaTeX headers (using raw strings 'r')
table_data = {
    r"Step ($n$)": n_values,
    r"Numerator ($n^2$)": numerators,
    r"Denominator ($2^n$)": denominators,
    r"Value ($\frac{n^2}{2^n}$)": np.round(results, 4), 
    r"Is $a_n < 0.5$?": results < 0.5           
}

df = pd.DataFrame(table_data)

Markdown(df.to_markdown(index=False))
```
